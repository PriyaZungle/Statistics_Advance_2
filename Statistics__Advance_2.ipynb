 1. What is hypothesis testing in statistics?
  --
 - An act in statistics whereby an analyst tests an assumption regarding a population parameter.
 - Hypothesis testing is a method used to make decisions or inferences about a population based on a sample of data.

 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
  --
 - The null hypothesis is a default assumption that there is no effect, no difference, or no relationship between variables.
 - The null hypothesis is the statement being tested.
 - Null Hypothesis:
    - To try to disprove
    - Based on default belief
    - Outcome if (p < 0.05)	Reject H₀
    - Outcome if (p ≥ 0.05)	Fail to reject H₀
 - Alternative Hypothesis
    - To try to support
    - Based on research or experimental claim
    - Outcome if p < 0.05 suggest evidence for H₁
    - Outcome if p ≥ 0.05 no strong evidence for H₁

3. What is the significance level in hypothesis testing, and why is it important?
 --
- In hypothesis testing, the significance level (alpha, denoted as α) is the probability of incorrectly rejecting the null hypothesis when it is actually true.
- Example : 0.01 → 1% chance etc.
- If your p-value ≤ α, you reject H₀ → Your result is statistically significant

- If your p-value > α, you fail to reject H₀ → Not enough evidence to support your claim.

4. What does a P-value represent in hypothesis testing?
 --
-  A p-value is a statistical measurement used to validate a hypothesis against observed data.
-  A p-value measures the probability of obtaining the observed results, assuming that the null hypothesis is true.
-The lower the p-value, the greater the statistical significance of the observed difference.

5. How do you interpret the P-value in hypothesis testing?
 --
- The p-value is the probability that the observed effect within the study would have occurred by chance if, in reality, there was no true effect.
- A (p < 0.05 )or (p < 0.01) is considered statistically significant.

6. What are Type 1 and Type 2 errors in hypothesis testing?
- Type I errors
  --
  - A Type I error is rejecting a true null hypothesis (false positive).
  - This occurs when you conclude that there is a significant effect or difference when, in reality, there isn't one.
  -  Type I error is denoted by alpha (α).
- Type II errors
  --
  -  Type II error is failing to reject a false null hypothesis (false negative).
  - This occurs when you fail to detect a significant effect or difference that actually exists.
   - Type II error is denoted by beta (β).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- One-Tailed Test
  --
  - Testing for a difference in one specific direction (greater than or less than).
  - It's used when the researcher has a strong prior belief about the direction of the effect.
  - H₀: New battery life ≤ Old battery life
  - H₁: New battery life > Old battery life →(Right-tailed)
- Two-Tailed Test:
  --
  - testing for any difference — it could be higher or lower.
  - It's used when the researcher is interested in detecting any deviation from the null hypothesis, regardless of direction.
  - H₀: New scores = Old scores
  - H₁: New scores ≠ Old scores →  (Two-tailed)

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_data = np.array(sample_data)
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    print("Sample Mean:", round(sample_mean, 4))
    print("Z-score:", round(z_score, 4))
    print("P-value:", round(p_value, 4))
    if p_value < alpha:
        print(f"Reject the null hypothesis at alpha = {alpha}.")
    else:
        print(f"Fail to reject the null hypothesis at alpha = {alpha}.")


people = [100, 104, 120, 109, 95, 99, 100, 100, 102, 79]
population_mean = 100
population_std = 2

z_test(people, population_mean, population_std)
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
np.random.seed(66)
population_mean = 75
population_std = 45
sample_size = 50
alpha = 0.05

sample_data = np.random.normal(loc=52, scale=population_std, size=sample_size)
sample_mean = np.mean(sample_data)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))
print("Sample Mean:", round(sample_mean, 4))
print("Z-score:", round(z_score, 4))
print("P-value:", round(p_value, 4))

if p_value < alpha:
    print(f"Reject the null hypothesis at alpha = {alpha}.")
else:
    print(f"Fail to reject the null hypothesis at alpha = {alpha}.")
z_vals = np.linspace(-4, 4, 1000)
pdf_vals = norm.pdf(z_vals)
z_critical = norm.ppf(1 - alpha / 2)
plt.figure(figsize=(10, 5))
plt.plot(z_vals, pdf_vals, label='Standard Normal Distribution', color='blue')
plt.fill_between(z_vals, 0, pdf_vals, where=(z_vals <= -z_critical), color='red', alpha=0.5, label='Rejection Region')
plt.fill_between(z_vals, 0, pdf_vals, where=(z_vals >= z_critical), color='red', alpha=0.5)
plt.axvline(z_score, color='green', linestyle='--', linewidth=2, label=f'Z-score = {z_score:.2f}')
plt.axvline(-z_critical, color='black', linestyle=':', label=f'Critical Z = ±{z_critical:.2f}')
plt.axvline(z_critical, color='black', linestyle=':')
plt.title('Two-Tailed Z-Test and Decision Region')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(true_mean, sample_size, std_dev, alpha=0.05, num_simulations=1000):
    critical_value = stats.norm.ppf(1 - alpha)
    type_1_errors = 0
    type_2_errors = 0
    for _ in range(num_simulations):
        sample_null = np.random.normal(0, std_dev, sample_size)
        sample_alternative = np.random.normal(true_mean, std_dev, sample_size)
        sample_mean_null = np.mean(sample_null)
        sample_mean_alt = np.mean(sample_alternative)
        z_null = (sample_mean_null) / (std_dev / np.sqrt(sample_size))
        if z_null > critical_value:
            type_1_errors += 1
        z_alt = (sample_mean_alt) / (std_dev / np.sqrt(sample_size))
        if z_alt < critical_value:
            type_2_errors += 1
    type_1_error_rate = type_1_errors / num_simulations
    type_2_error_rate = type_2_errors / num_simulations
    print(f"Type 1 Error Rate (False Positive): {type_1_error_rate:.4f}")
    print(f"Type 2 Error Rate (False Negative): {type_2_error_rate:.4f}")
    labels = ['Type 1 Error', 'Type 2 Error']
    error_rates = [type_1_error_rate, type_2_error_rate]

    plt.bar(labels, error_rates, color=['red', 'black'])
    plt.xlabel('Error Type')
    plt.ylabel('Error Rate')
    plt.title('Type 1 and Type 2 Error Rates')
    plt.ylim(0, 1)
    plt.show()
true_mean = 2
sample_size = 50
std_dev = 1

visualize_errors(true_mean, sample_size, std_dev, alpha=0.05, num_simulations=1000)
```

```python

#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    df = n - 1
    t_critical = stats.t.ppf(1 - (1 - confidence) / 2, df)
    standard_error = std_dev / np.sqrt(n)
    margin_of_error = t_critical * standard_error
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return (lower_bound, upper_bound)
sample_data = np.random.normal(66, 44, 88)
ci = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval: {ci}")
```

```python
#10 Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    n = len(data)
    std_dev = np.std(data, ddof=1)
    df = n - 1
    t_critical = stats.t.ppf(1 - (1 - confidence) / 2, df)
    standard_error = std_dev / np.sqrt(n)
    margin_of_error = t_critical * standard_error

    return margin_of_error

sample_data = np.random.normal(66, 44, 88)
confidence_level = 0.95
moe = margin_of_error(sample_data, confidence=confidence_level)
print(f"Margin of Error at {confidence_level*100}% confidence level: {moe:.3f}")
```

```python
#13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.


import numpy as np

def calculate_expected_frequencies(observed):
    observed = np.array(observed)
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    grand_total = observed.sum()
    expected = np.zeros_like(observed, dtype=float)
    for i in range(observed.shape[0]):
        for j in range(observed.shape[1]):
            expected[i, j] = (row_totals[i] * col_totals[j]) / grand_total

    return expected
observed_data = [
    [40, 30],
    [70, 40]
]
expected_frequencies = calculate_expected_frequencies(observed_data)
print("Observed Frequencies:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python

#16 Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    # Step 1: Calculate sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    if var1 < var2:
        var1, var2 = var2, var1
    F = var1 / var2
    dfn = len(sample1) - 1
    dfd = len(sample2) - 1
    f_critical_low = f.ppf(alpha / 2, dfn, dfd)
    f_critical_high = f.ppf(1 - alpha / 2, dfn, dfd)
    p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))
    print(f"Sample 1 Variance: {var1:.4f}")
    print(f"Sample 2 Variance: {var2:.4f}")
    print(f"F-statistic: {F:.4f}")
    print(f"Degrees of Freedom (df1, df2): ({dfn}, {dfd})")
    print(f"Critical Values (low, high): ({f_critical_low:.4f}, {f_critical_high:.4f})")
    print(f"P-value: {p_value:.4f}")
    if F < f_critical_low or F > f_critical_high:
        print("Conclusion: Reject the null hypothesis. Variances are significantly different.")
    else:
        print("Conclusion: Fail to reject the null hypothesis. No significant difference in variances.")
np.random.seed(0)
sample1 = np.random.normal(loc=9, scale=3, size=30)
sample2 = np.random.normal(loc=10, scale=2, size=30)

f_test(sample1, sample2)
```

```python
#19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
import numpy as np
from scipy.stats import shapiro, levene
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(*groups):

    all_data = []
    labels = []
    for i, group in enumerate(groups):
        stat, p = shapiro(group)
        all_data.extend(group)
        labels.extend([f'Group {i+1}'] * len(group))
        print(f"Group {i+1} Normality (Shapiro-Wilk): p-value = {p:.4f}")
        if p > 0.05:
            print("  ✅ Normal distribution (fail to reject H₀)")
        else:
            print("  ❌ Not normal (reject H₀)")
    stat, p = levene(*groups)
    print(f"\nEqual Variance (Levene's Test): p-value = {p:.4f}")
    if p > 0.05:
        print("  ✅ Equal variances (fail to reject H₀)")
    else:
        print("  ❌ Unequal variances (reject H₀)")
    plt.figure(figsize=(4, 9))
    sns.boxplot(x=labels, y=all_data, palette="light:#5A9")
    plt.title("Boxplot of Groups for Visual Inspection")
    plt.ylabel("Values")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()
np.random.seed(0)
group_1 = np.random.normal(loc=20, scale=5, size=30)
group_2 = np.random.normal(loc=22, scale=5, size=30)
group_3 = np.random.normal(loc=24, scale=5, size=30)
check_anova_assumptions(group_1, group_2, group_3)
```

```python
#20 Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(1)
factor_A = np.repeat(['Low', 'High'], 30)
factor_B = np.tile(np.repeat(['Control', 'Treatment'], 15), 2)
data = pd.DataFrame({
    'A': factor_A,
    'B': factor_B,
    'Score': np.random.normal(70, 5, 60) +
             (np.array(factor_A) == 'High') * 5 +
             (np.array(factor_B) == 'Treatment') * 3 +
             ((np.array(factor_A) == 'High') & (np.array(factor_B) == 'Treatment')) * 4
})
model = ols('Score ~ C(A) * C(B)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("Two-way ANOVA Table:")
print(anova_table)
plt.figure(figsize=(8, 5))
sns.pointplot(data=data, x='A', y='Score', hue='B',
              dodge=True, markers=['o', 's'], capsize=.1, errwidth=1.5, palette='Set2')
plt.title('Interaction Plot: Factor A vs. Factor B')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#22 Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway
import pandas as pd

np.random.seed(42)
group_A = np.random.normal(loc=50, scale=5, size=30)
group_B = np.random.normal(loc=55, scale=5, size=30)
group_C = np.random.normal(loc=60, scale=5, size=30)

data = pd.DataFrame({
    'Score': np.concatenate([group_A, group_B, group_C]),
    'Group': ['A']*30 + ['B']*30 + ['C']*30
})

f_stat, p_value = f_oneway(group_A, group_B, group_C)
print("One-way ANOVA Results")
print("----------------------")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Conclusion: Reject the null hypothesis. At least one group mean is different.")
else:
    print("Conclusion: Fail to reject the null hypothesis. No significant difference between group means.")
plt.figure(figsize=(6, 9))
sns.boxplot(x='Group', y='Score', data=data, palette='Set2')
plt.title('Comparison of Group Means (Boxplot)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#24 Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
import numpy as np
import scipy.stats as stats


data = [5.3,5.2,5.5,6.1,6.9,5.4,5.1,5.2,6.3,6.8,6.2,8.7]
n = len(data)
sample_variance = np.var(data, ddof=1)
sigma_0_squared = 4
chi_square_stat = (n - 1) * sample_variance / sigma_0_squared
df = n - 1
alpha = 0.05
chi_square_critical_low = stats.chi2.ppf(alpha / 2, df)
chi_square_critical_high = stats.chi2.ppf(1 - alpha / 2, df)
if chi_square_stat < chi_square_critical_low or chi_square_stat > chi_square_critical_high:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"
print(f"Test Statistic (Chi-square): {chi_square_stat}")
print(f"Critical Values: ({chi_square_critical_low}, {chi_square_critical_high})")
print(f"Decision: {decision}")
```

```python
#25import numpy as np
import scipy.stats as stats

def z_test_proportions(success_a, sample_size_a, success_b, sample_size_b):
    # Calculate the sample proportions
    p_hat_a = success_a / sample_size_a
    p_hat_b = success_b / sample_size_b
    p_pool = (success_a + success_b) / (sample_size_a + sample_size_b)
    se = np.sqrt(p_pool * (1 - p_pool) * (1 / sample_size_a + 1 / sample_size_b))
    z_statistic = (p_hat_a - p_hat_b) / se
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
    print(f"Z Statistic: {z_statistic}")
    print(f"P-value: {p_value}")
    return z_statistic, p_value
success_a = 60
sample_size_a = 100
success_b = 45
sample_size_b = 100
z_stat, p_val = z_test_proportions(success_a, sample_size_a, success_b, sample_size_b)

alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: There is a significant difference between the proportions.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the proportions.")
```

```python
#26 Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def f_test_variance(data_a, data_b):

    var_a = np.var(data_a, ddof=1)
    var_b = np.var(data_b, ddof=1)


    f_statistic = var_a / var_b if var_a > var_b else var_b / var_a

    # Degrees of freedom
    df_a = len(data_a) - 1
    df_b = len(data_b) - 1

    # Perform the F-test to get the p-value
    p_value = 2 * min(stats.f.cdf(f_statistic, df_a, df_b), 1 - stats.f.cdf(f_statistic, df_a, df_b))

    # Print results
    print(f"F-statistic: {f_statistic}")
    print(f"P-value: {p_value}")

    return f_statistic, p_value

# Example usage
data_a = np.random.normal(loc=70, scale=10, size=140)
data_b = np.random.normal(loc=90, scale=15, size=180)

f_stat, p_val = f_test_variance(data_a, data_b)
alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
plt.figure(figsize=(10, 6))
plt.hist(data_a, alpha=0.5, label='Group A', bins=20, color='blue')
plt.hist(data_b, alpha=0.5, label='Group B', bins=20, color='red')
plt.legend()
plt.title("Histogram of Data Groups A and B")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#27 Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
np.random.seed(42)
observed_data = np.random.choice([1, 2, 3, 4, 5], size=1000, p=[0.1, 0.2, 0.3, 0.2, 0.2])
observed_frequencies = [np.sum(observed_data == i) for i in range(1, 6)]
expected_frequencies = [200] * 5
chi2_stat, p_val = stats.chisquare(observed_frequencies, expected_frequencies)
print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_val}")
alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: The observed frequencies do not follow the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed frequencies follow the expected distribution.")
labels = ['1', '2', '3', '4', '5']
x = np.arange(len(labels))

plt.bar(x - 0.2, observed_frequencies, 0.4, label='Observed', color='blue')
plt.bar(x + 0.2, expected_frequencies, 0.4, label='Expected', color='red')
plt.xticks(x, labels)
plt.xlabel("Category")
plt.ylabel("Frequency")
plt.title("Chi-square Goodness of Fit Test\nObserved vs Expected Frequencies")
plt.legend()
plt.show()
```
